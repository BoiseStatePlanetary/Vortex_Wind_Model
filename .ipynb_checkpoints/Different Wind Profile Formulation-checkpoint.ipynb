```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def wind_profile(t, Vobs, Gamma_wind, Gamma_obs):
    
    if(Gamma_wind == 0):
        return Vobs*2*(t/Gamma_obs)/(1. + (t/Gamma_obs)**2)
    else:
        return Vobs*np.sqrt(1 + (t/Gamma_wind)**2)/(1 + (2.*t/Gamma_obs)**2)

Gamma_obs = 30. # seconds
Dact = 10 # meters
b = 0.
Vobs = 10. # m/s

Dobs = np.sqrt((2*b)**2 + Dact**2)

Gamma_wind = Gamma_obs*b/Dobs

t = np.linspace(-5*Gamma_obs, 5.*Gamma_obs, 1000)
Vr = wind_profile(t, Vobs, Gamma_wind, Gamma_obs)

plt.plot(t, Vr)
```
