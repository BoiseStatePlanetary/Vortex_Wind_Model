Figure 4 from [Franzese et al. (2020)](https://ui.adsabs.harvard.edu/abs/2020AeoRe..4400594F/abstract).

The radial distance between the wind sensor and the vortex center $d$ is given by 
$$
d = \sqrt{s^2t^2 + d_0^2},
$$
where $s$ is the magnitude of the background wind, $t$ is the time from closest approach (and runs from $-\infty$ to $+\infty$, with the closest approach occurring at $t = 0$), and $d_0$ is the closest approach distance.

The $x$ and $y$ components of the displacement vector $\vec{d}$ between the sensor and vortex center are given by 
$$
d_{x} = s \cos \left( \theta_s \right)\ t + d_{0x}\\
d_{y} = s \sin \left( \theta_s \right)\ t + d_{0y}.
$$
where $\theta_s$ is the angle between north ($x$) and the background wind vector $\vec{s}$ and $d_{0x/y}$ is the $x/y$ displacement at closest approach $t = 0$. In the figure above, $\theta_s = 0$, $d_{0x} = 0$, $d_{0y} = d_0$.

The total wind vector measured at the sensor $\vec{v}_{\rm t}$ is the vector sum of the background wind $\vec{s}$ and the rotational wind associated with the vortex, $\vec{v}_{\rm r}$: 
$$
\vec{v}_{\rm t} = \vec{s} + \vec{v}_{\rm r}.
$$

The vortex's wind vector is a function of the displacement vector $\vec{d}$, i.e. $\vec{v}_{\rm r} = \vec{v}_{\rm r}(\vec{d})$.
